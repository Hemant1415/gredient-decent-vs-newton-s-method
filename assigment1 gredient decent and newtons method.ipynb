```python
import numpy as np
```

```python
# GRADIENT DESCENT METHOD
import sympy
list_step=[]
list_func=[]  # these lists are defined to create barh
list_functions=[]
```

```python
from sympy import *
x,y=symbols('x y')
def decent(func,x_0,step_size,tolerance,colour): # INPUTS REQUIRED ARE FUNC,INTIAL POINT, STEP SIZE TOLERANCE AND COLOUR OF GRAPH
    diff=Derivative(func,x).doit()
    x_1=x_0-diff.subs(x,x_0)*step_size
    list2=[x_0,x_1]
    list1=[0,1]
    count=1
    title=str(func)+" with step size "+str(step_size)+" and tolerance "+str(tolerance)+" and initial point "+str(x_0)
    list_functions.append(func)
    while abs(x_1-x_0)>tolerance:
        x_1,x_0=(x_1-(diff.subs(x,x_1))*step_size),x_1
        list2.append(x_1)
        count=count+1
        list1.append(count)
    plt.plot(list1,list2,marker="o",color=colour,label=title)
    plt.title(str(func))
    plt.xlabel("no of iteration")
    plt.ylabel("value of x_n")
    plt.grid()
    plt.legend(bbox_to_anchor=(1.44,-0.25),loc='upper right')
    plt.tight_layout()
    list_step.append(count)
    list_func.append(title)
```

```python
from sympy import*
x,y=symbols('x y')
decent(x**2+6*x+3,4,0.1,0.1,"red")
decent(x**2+6*x+3,5,0.2,0.01,"yellow")
decent(x**2+6*x+3,6,0.3,0.001,"green")
plt.show()
decent(exp(x)-log(x),1,0.1,0.1,"blue")
decent(exp(x)-log(x),2,0.2,0.1,"orange")
decent(exp(x)-log(x),3,0.3,0.1,"black")
plt.show()
decent(x**2+8*x-2,2,0.1,0.1,"black")
decent(x**2+8*x-2,3,0.2,0.01,"blue")
decent(x**2+8*x-2,4,0.3,0.001,"violet")
plt.show()
# CALL THE FUNCTION DECENT HERE GIVE it ALL INPUTS REQUIRED 
```

```python
from sympy import*
x,y=symbols('x y')
def newton(func,x_0,tolerance,colour): # INPUTS REQUIRED ARE FUNC,INTIAL POINT, STEP SIZE TOLERANCE AND COLOUR OF GRAPH
    diff=Derivative(func,x).doit()
    x_1=x_0-func.subs(x,x_0)/diff.subs(x,x_0)
    list2=[x_0,x_1]
    list1=[0,1]
    count=1
    title=str(func )+" with tolerance "+str(tolerance)+" and initial point "+str(x_0)
    while abs(x_1-x_0)>tolerance:
        x_1,x_0=float(x_1-(func.subs(x,x_1))/diff.subs(x,x_1)),x_1
        list2.append(x_1)
        count=count+1
        list1.append(count)
    
    plt.plot(list1,list2,marker="o",color=colour,label=title)
    plt.title(str(func))
    plt.xlabel("no of iteration")
    plt.ylabel("value of x_n")
    plt.grid()
    plt.legend(bbox_to_anchor=(1.44,-0.25),loc='upper right')
    plt.tight_layout()
    list_steps.append(count)
    list_func2.append(title)
```

```python
from sympy import*
x,y=symbols('x y')
newton(x**2+6*x+3,4,0.01,"red")
newton(x**2+6*x+3,5,0.01,"yellow")
newton(x**2+6*x+3,6,0.01,"green")
plt.show()
newton(x**2+8*x-2,2,0.01,"black")
newton(x**2+8*x-2,3,0.01,"blue")
newton(x**2+8*x-2,4,0.01,"violet")
plt.show()
newton(exp(x)-log(x),5,1,"blue")
newton(exp(x)-log(x),6,1,"orange")
newton(exp(x)-log(x),7,1,"black")
plt.show()
```

```python

categories = list_func
group1_values = list_steps
group2_values = list_step

bar_height = 0.35
bar_positions_group1 = np.arange(len(categories))
bar_positions_group2 = bar_positions_group1 + bar_height

plt.barh(bar_positions_group1, group1_values, height=bar_height, label='newtons method')
plt.barh(bar_positions_group2, group2_values, height=bar_height, label='gradient decent')

plt.xlabel('no of iterations')
plt.ylabel('function ')
plt.title('Grouped Horizontal Bar Chart')
plt.yticks(bar_positions_group1 + bar_height / 2, categories)
plt.legend()

plt.show()
```
